## **GTO Poker Regret Minimization Demo Notebook**

```python
from cfr_solver import CFRSolver
import matplotlib.pyplot as plt

solver = CFRSolver(num_iterations=10000)
solver.train()
avg_strategy = solver.get_average_strategy()

# Plot
import seaborn as sns
import pandas as pd

hands = solver.hands
actions = solver.actions
df = pd.DataFrame(avg_strategy, columns=actions, index=hands)
sns.heatmap(df, annot=True, cmap="YlGnBu", cbar=False)
plt.title("Average Strategy pesr Hand (Toy GTO Solver)")
plt.show()
```
